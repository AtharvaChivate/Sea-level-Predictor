```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
df = pd.read_csv('epa-sea-level.csv')
```

```python
df.head()
```

```python
# Creating a scatter plot
y=df['CSIRO Adjusted Sea Level']
x=df['Year']

fig, ax = plt.subplots()
ax.scatter(x,y)
```

```python
# Create first line of best fit
res = linregress(df['Year'], df['CSIRO Adjusted Sea Level'])
x_pred = pd.Series([i for i in range(1880, 2051)])
y_pred = res.slope*x_pred + res.intercept

plt.plot(x_pred, y_pred, "r")
```

```python
# Create second line of best fit
new_df = df.loc[df['Year'] >= 2000]
new_x = new_df['Year']
new_y = new_df['CSIRO Adjusted Sea Level']

res_2 = linregress(new_x, new_y)
x_pred2 = pd.Series([i for i in range(2000, 2051)])
y_pred2 = res_2.slope*x_pred2 + res_2.intercept

plt.plot(x_pred2, y_pred2, "green")
```

```python
# Adding labels and title
ax.set_xlabel('Year')
ax.set_ylabel('Sea Level (inches)')
ax.set_title('Rise in Sea Level')

plt.savefig('sea_level_plot.png')
```

```python
# Creating a scatter plot
y = df['CSIRO Adjusted Sea Level']
x = df['Year']

fig, ax = plt.subplots()
ax.scatter(x, y)

# Creating first line of best fit
res = linregress(df['Year'], df['CSIRO Adjusted Sea Level'])
x_pred = pd.Series([i for i in range(1880, 2051)])
y_pred = res.slope * x_pred + res.intercept

# Plotting the first line of best fit
ax.plot(x_pred, y_pred, "r", label='Best Fit Line 1880-2050')

# Creating second line of best fit
new_df = df.loc[df['Year'] >= 2000]
new_x = new_df['Year']
new_y = new_df['CSIRO Adjusted Sea Level']

res_2 = linregress(new_x, new_y)
x_pred2 = pd.Series([i for i in range(2000, 2051)])
y_pred2 = res_2.slope * x_pred2 + res_2.intercept

# Plotting the second line of best fit
ax.plot(x_pred2, y_pred2, "green", label='Best Fit Line 2000-2050')

# Setting labels and title
ax.set_xlabel('Year')
ax.set_ylabel('CSIRO Adjusted Sea Level')
ax.set_title('Sea Level Predictions')

# Adding legend
ax.legend()

plt.savefig('sea_level_plot.png')

# Display the plot
plt.show()
```
